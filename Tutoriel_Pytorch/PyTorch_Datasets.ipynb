**Datasets et DataLoader**

Nomenclature :
- epoch : 1 foward and backward pass of ALL training sample
- batch_size : number of training samples in one foward & backward pass
- number of iteration : number of passes, each pass using [batch_size] number of sample

ex: 100 samples, batch_size = 20 -> 100/20 = 5 iterations for 1 epoch

```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
import numpy as np
import math

class WineDatasets(Dataset):

    def __init__(self):
        xy = np.loadtxt("wine.txt", dtype=np.float32, skiprows=1, delimiter=",")
        self.x = torch.from_numpy(xy[:,1:])
        self.y = torch.from_numpy(xy[:,[0]])  #n_sample, 1
        self.n_sample = xy.shape[0]

    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def __len__(self):
        return self.n_sample

dataset = WineDatasets()
#first_data = dataset[0]
#features, labels = first_data
#print(features, first_data)

dataloader = DataLoader(dataset=dataset, batch_size=4, shuffle=True, num_workers=0)
#num_workers accelerate the calculation -> peut faire crash le code


datatiter = iter(dataloader)  #On transforme le data en iterable
data = datatiter.next()
features, labels = data

print(features, labels)

#On constate 4 "feature" vecteur car on a demand que batch_size=4

#Training Loop (plutt que simplement le next)

n_iters = 100
total_samples = len(dataset)
n_iteration = math.ceil(total_samples/4)  #Arrondi le nombre  la hausse

for epoch in range(n_iters):
    for i, (inputs, labels) in enumerate(dataloader):
        #foward backward, update
        if (i+1) % 5 == 0:
            print(f"epoch : {i+1}/{n_iters}, step {i+1}/{n_iteration}, input: {inputs.shape}")
```

**Transformer le dataset en tenseur avec les transformations**

```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
import numpy as np
import math

class WineDatasets(Dataset):

    def __init__(self, transform=None):
        xy = np.loadtxt("wine.txt", dtype=np.float32, skiprows=1, delimiter=",")
        self.x = torch.from_numpy(xy[:,1:])
        self.y = torch.from_numpy(xy[:,[0]])  #n_sample, 1
        self.n_sample = xy.shape[0]

        self.transform = transform

    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def __len__(self):
        return self.n_sample


class ToTensor:

    def __call__(self, sample):  #Permet d'agir comme une fonction
        inputs, targets = sample
        return torch.from_numpy(inputs), torch.from_numpy(targets)

dataset = WineDatasets(transform=ToTensor())

first_data = dataset[0]
feature, label = first_data
print(type(feature), type(label))  #Si transform=ToTensor() -> torch.Tensor autrement np.array
```
