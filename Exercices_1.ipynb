**1 Fonction sigmoïde**

Soit la fonction sigmoïde $\sigma$ :

$\mathbb{R} \rightarrow \mathbb{R}$

$x \mapsto \sigma(x) = (1+e^{-x})^{-1}$

*a) Démontrez que $0<\sigma(x)<1$ pour tout $x$.  C'est donc dire que l'image de $\sigma$ est l'intervalle ouvert $(0,1)$:*


Il sera facile de montrer numériquement que cette fonction sigmoide a un comportement asymptotique à 0 et 1. Toutefois, on peut rapidement s'en convaincre mathématiquement. Premièrement, on constate que la fonction ne possède pas de discontinuité. Elle est continue en tout point. En effectuant la dérivé :

$\frac{d\sigma(x)}{dx} = \frac{e^{-x}}{(e^{-x}+1)^2}$

On constate que la dérivé est toujours positive. En effet, $e^{-x}$ est toujours positif et le dénominateur est au carré. C'est donc une équation monotone croissante. Ainsi, puisque la fonction est continu, nous savons que le minimum est à $-\infty$ et le maximum à $\infty$. On trouve ces valeurs frontières en calculant la limite :

$\lim_{x \to -\infty} \frac{1}{1+e^{-x}} = 0$

$lim_{x \to \infty} \frac{1}{1+e^{-x}} = 1$

On vient donc de trouver que le domaine est $[0,1]$. On le démontre numériquement :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = lambda x: (1+np.exp(-x))**-1

x1 = np.linspace(-1000, 1000, 100000)
x2 = np.linspace(-10, 10, 100000)

plt.figure(1)

plt.subplot(121)
plt.plot(x1, sigma(x1), "r-")
plt.title(r"$\sigma(x)$ pour x $\in [-1000, 1000]$")
plt.xlabel(r"$x$")
plt.ylabel(r"$\sigma(x)$")
plt.grid(True)
plt.subplot(122)
plt.grid(True)
plt.xlabel(r"$x$")
plt.ylabel(r"$\sigma(x)$")
plt.plot(x2, sigma(x2), "b-")
plt.grid(True)
plt.title(r"$\sigma(x)$ pour x $\in [-10, 10]$")
plt.show()

# On se convaint que le domaine est bien entre 0 et 1 aux extrèmes :
print(f"à -infini : {sigma(-np.inf)}")
print(f" à +infini : {sigma(np.inf)}")
```

On voit clairement, numériquement, que le domaine est entre 0 et 1.

*f) Démontrez que si $\alpha\leq 4$, alors l'équation possède un seul point d'équilibre.  Quelle est la stabilité linéaire de ce point d'équilibre?*

Tout d'abord, debutons par dessiner la fonction à $\alpha = 3$, $\alpha = 4$ et $\alpha = 5$ pour avoir une idée de son comportement. Arbitraiement, on utilise $\beta = 2$

```python
def y_dot(alpha, beta, y):
    return -y - beta + alpha * ((1+np.exp(-y))**(-1))

y = np.linspace(-3, 3, 1000)
plt.figure(1)

plt.subplot(131)
plt.tight_layout()
plt.plot(y, y_dot(3, 2, y), "r-")
plt.xlabel(r"$y$")
plt.ylabel(r"$\dot{y}$")
plt.title(r"$\dot{y}$ avec $\alpha = 3$")

plt.subplot(132)
plt.tight_layout()
plt.plot(y, y_dot(4, 2, y), "r-")
plt.xlabel(r"$y$")
plt.ylabel(r"$\dot{y}$")
plt.title(r"$\dot{y}$ avec $\alpha = 4$")

plt.subplot(133)
plt.tight_layout()
plt.plot(y, y_dot(5, 2, y), "r-")
plt.xlabel(r"$y$")
plt.ylabel(r"$\dot{y}$")
plt.title(r"$\dot{y}$ avec $\alpha = 5$")

plt.show()
```

Rapidement, ce qu'on semble remarquer, c'est que la fonction est monotone décroissante pour $\alpha \leq 4$. Ainsi, la fonction aura toujours qu'un seul point d'équilibre puisque l'on obtient $\dot{y}$ une seule fois. On peut s'en convaincre en effectuant la dérivée :

$\ddot{y} = -1 + \alpha \cdot \frac{e^{-y}}{(1+e^{-y})^2}$

On constate que le seul paramètre que l'on peut modifier est $\alpha$. On représente graphiquement la fonction :

```python
def y_ddot(alpha, y):
    return -1 + alpha * np.exp(-y)/(1+np.exp(-y))**2

y = np.linspace(-5, 5, 1000)

plt.plot(y, y_ddot(4, y), "r-", label=r"$\alpha = 4$")
plt.plot(y, y_ddot(3, y), "b-", label=r"$\alpha = 3$")
plt.plot(y, y_ddot(5, y), "g-", label=r"$\alpha=5$")
plt.grid(True)
plt.legend()
plt.xlabel(r"$y$")
plt.ylabel(r"$\ddot{y}$")
plt.axhline(0)
plt.show()
```

Ce que ce graphique nous indique, c'est que la dérivée de la fonction $\dot{y}$ est toujours négative tant et aussi longtemps que $\alpha \leq 4$. Dans ce cas-ci, nous avons une fonction continue monotone décroissante. Dans le cas où $\alpha > 4$, alors il se trouve des endroits où la dérivé de $\dot{y}$ est positive. Il est donc possible que la fonction $\dot{y}$ passe plus d'une fois sur l'axe des abscisses et donc la fonction peut posséder plus d'un point d'équilibre.

Un point d'équilibre est dit stable si $\frac{dF}{dx}|_{x=x_0} < 0$ et instable si $\frac{dF}{dx}|_{x=x_0} > 0$. On constate que lorsque $\alpha \leq 4$, alors la dérivée est toujours négative (monotone décroissante). C'est donc dire que l'équilibre est stable. (**à confirmer**)
